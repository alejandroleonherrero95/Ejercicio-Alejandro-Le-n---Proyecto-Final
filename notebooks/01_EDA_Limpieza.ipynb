```python
import pandas as pd

# Cargar los datasets (con el nombre exacto que devuelve os.listdir)
job_postings = pd.read_csv("../data/raw/job_postings_europe.csv")
employment = pd.read_csv("../data/raw/employment_eurostat.tsv.gz.", sep="\t")

# Verificar tamaños
print("Job postings shape:", job_postings.shape)
print("Employment shape:", employment.shape)

# Mostrar primeras filas
display(job_postings.head())
display(employment.head())
```

```python
# Información general
print("Job postings info:")
display(job_postings.info())

print("\nEmployment info:")
display(employment.info())
```

```python
# --- Limpieza inicial ---

# Valores nulos por columna
print("Valores nulos en Job Postings:")
display(job_postings.isnull().sum())

print("\nValores nulos en Employment:")
display(employment.isnull().sum())

# Filas duplicadas
print("\nDuplicados en Job Postings:", job_postings.duplicated().sum())
print("Duplicados en Employment:", employment.duplicated().sum())

# Distribución rápida de columnas principales (Job postings)
print("\nDistribución de categorías de empleo:")
display(job_postings['category'].value_counts().head())

print("\nDistribución de localizaciones:")
display(job_postings['location'].value_counts().head())
```

```python
# --- Limpieza de datos ---

# 1. Eliminar duplicados en job_postings
job_postings_clean = job_postings.drop_duplicates()
print("Tamaño original:", job_postings.shape)
print("Tamaño tras eliminar duplicados:", job_postings_clean.shape)

# 2. Manejo de valores nulos
# - Eliminamos filas con todos los valores nulos
job_postings_clean = job_postings_clean.dropna(how="all")

# - Para salary_offered, rellenamos con la mediana (opcional)
job_postings_clean["salary_offered"] = job_postings_clean["salary_offered"].fillna(
    job_postings_clean["salary_offered"].median()
)

print("\nValores nulos tras limpieza:")
print(job_postings_clean.isnull().sum())

# Revisamos nuevamente las primeras filas
display(job_postings_clean.head())
```

```python
# --- Limpieza de Job Postings ---
job_postings_clean = job_postings.copy()

# Eliminar filas completamente vacías
job_postings_clean = job_postings_clean.dropna(how="all")

# Rellenar nulos de salary_offered con "Unknown"
job_postings_clean["salary_offered"] = job_postings_clean["salary_offered"].fillna("Unknown")

print("\nValores nulos tras limpieza:")
print(job_postings_clean.isnull().sum())

# Mostrar algunas filas para verificar
display(job_postings_clean.head())
```

```python
# ---- Limpieza de Employment ----
employment_clean = employment.copy()

# Separar la columna combinada en varias
employment_clean = employment_clean.rename(columns={
    "freq,indic_em,sex,age,unit,geo\\TIME_PERIOD": "meta"
})

# Separar en columnas nuevas
employment_clean[["freq","indic_em","sex","age","unit","geo"]] = (
    employment_clean["meta"].str.split(",", expand=True)
)

# Eliminar columna original
employment_clean = employment_clean.drop(columns=["meta"])

# Convertir años a numéricos (si es posible)
for col in employment_clean.columns:
    if col.isdigit():  # si es un año
        employment_clean[col] = pd.to_numeric(employment_clean[col], errors="coerce")

print("Valores nulos tras limpieza en Employment:")
print(employment_clean.isnull().sum().sum())

display(employment_clean.head())
```

```python
# ---- Normalizar fechas en Job Postings ----
job_postings_clean["post_date"] = pd.to_datetime(job_postings_clean["post_date"], errors="coerce")
job_postings_clean["year"] = job_postings_clean["post_date"].dt.year

print("Job postings con columna year:")
display(job_postings_clean[["post_date", "year"]].head())

# ---- Pivotar Employment ----
employment_long = employment_clean.melt(
    id_vars=["freq","indic_em","sex","age","unit","geo"],
    var_name="year",
    value_name="employment_value"
)

# Convertir year a numérico
employment_long["year"] = pd.to_numeric(employment_long["year"], errors="coerce")

print("Employment en formato largo:")
display(employment_long.head())
```

```python
# ---- Paso 5: Verificación antes de fusionar ----

# Años disponibles en cada dataset
print("Años en Job Postings:", sorted(job_postings_clean["year"].unique()))
print("Años en Employment:", sorted(employment_long["year"].unique()))

# Ubicaciones en Job Postings
print("\nUbicaciones únicas en Job Postings:")
print(job_postings_clean["location"].unique()[:20])  # mostramos solo primeros 20

# Ubicaciones en Employment
print("\nCódigos geo en Employment:")
print(employment_long["geo"].unique()[:20])  # mostramos solo primeros 20
```

```python
# Diccionario de mapeo manual
location_to_geo = {
    "UK": "UK",
    "London": "UK",
    "Cambridge": "UK",
    "M4 Corridor": "UK",
    "South East": "UK",
    "Manchester": "UK",
    "Birmingham": "UK",
    "North West": "UK",
    "Scotland": "UK",
    "Oxford": "UK",

    "Germany": "DE",
    "germany": "DE",

    "France": "FR",
    "Paris": "FR",

    "Switzerland": "CH",
    "switzerland": "CH",
    "Basel": "CH",

    "Spain": "ES",
    "Italy": "IT",
    "Europe": "EU27_2020",  # conjunto Unión Europea
}

# Crear columna geo en Job Postings
job_postings_clean["geo"] = job_postings_clean["location"].map(location_to_geo)

# Verificar valores sin asignar
print("Valores no mapeados:")
print(job_postings_clean[job_postings_clean["geo"].isna()]["location"].unique())
```

```python
# ---- Normalizar y mapear ubicaciones ----

# Normalizar ubicaciones a formato Title Case (primera mayúscula, resto minúsculas)
job_postings_clean["location_norm"] = job_postings_clean["location"].str.strip().str.title()

# Mapear usando la columna normalizada
job_postings_clean["geo"] = job_postings_clean["location_norm"].map(location_to_geo)

# Verificar valores no mapeados
print("Valores no mapeados después de normalizar:")
print(job_postings_clean[job_postings_clean["geo"].isna()]["location_norm"].unique())
```

```python
# ---- Ajustar diccionario con los últimos valores ----

# Añadimos las claves que faltaban
location_to_geo.update({
    "Uk": "UK",         # corresponde al Reino Unido
    "Portugal": "PT"    # código ISO para Portugal
})

# Reaplicamos el mapeo
job_postings_clean["geo"] = job_postings_clean["location_norm"].map(location_to_geo)

# Verificar que ya no quedan valores sin asignar
print("Valores no mapeados tras la corrección final:")
print(job_postings_clean[job_postings_clean["geo"].isna()]["location_norm"].unique())
```

```python
# ---- Guardar datasets limpios ----

# Guardar Job Postings limpio
job_postings_clean.to_csv("../data/job_postings_clean.csv", index=False)

# Guardar Employment limpio
employment_clean.to_csv("../data/employment_clean.csv", index=False)

print("✅ Archivos limpios guardados en la carpeta data/")
```

```python
# ---- Reducir dataset fusionado a ~80.000 filas ----

# Número total deseado
target_rows = 80000

# Calculamos el tamaño por grupo (geo, year)
fusionado_reducido = (
    fusionado.groupby(["geo", "year"], group_keys=False)
    .apply(lambda x: x.sample(
        n=min(len(x), max(1, target_rows // fusionado.groupby(["geo", "year"]).ngroups)), 
        random_state=42
    ))
)

# Verificamos resultado
print("Tamaño final del dataset reducido:", fusionado_reducido.shape)
print("Distribución por geo:")
print(fusionado_reducido["geo"].value_counts().head())

# Guardar versión reducida
fusionado_reducido.to_csv("../data/fusionado_reducido.csv", index=False)
print("✅ Dataset reducido guardado en data/fusionado_reducido.csv")
```
