```python
# ============================================================
# 03_EDA_Avanzado.ipynb — Exploración Avanzada de Datos (EDA)
# ============================================================

# 1️⃣ Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuración general
sns.set(style="whitegrid", palette="muted", font_scale=1.1)
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["axes.labelsize"] = 12

print("✅ Librerías importadas correctamente.")


# 2️⃣ Carga del dataset
df = pd.read_csv("../data/fusionado_reducido.csv")
print(f"\n📊 Dataset cargado correctamente con {df.shape[0]} filas y {df.shape[1]} columnas.\n")

# Vista previa
display(df.head())


# 3️⃣ Análisis general del dataset
print("----- INFORMACIÓN GENERAL -----")
df.info()

print("\n----- VALORES NULOS -----")
print(df.isnull().sum().sort_values(ascending=False).head(10))

print("\n----- ESTADÍSTICAS DESCRIPTIVAS NUMÉRICAS -----")
display(df.describe(include=[np.number]).T)

print("\n----- ESTADÍSTICAS DESCRIPTIVAS CUALITATIVAS -----")
display(df.describe(include=['object']).T)


# 4️⃣ Distribución de variables numéricas
numeric_cols = df.select_dtypes(include=[np.number]).columns

if len(numeric_cols) > 0:
    df[numeric_cols].hist(bins=25, figsize=(14, 10), color="skyblue", edgecolor="black")
    plt.suptitle("Distribución de variables numéricas", fontsize=16)
    plt.show()
else:
    print("⚠️ No se detectaron columnas numéricas para graficar.")


# 5️⃣ Detección y análisis de outliers
if len(numeric_cols) > 0:
    print("\n----- DETECCIÓN DE OUTLIERS (usando Z-Score) -----")
    z_scores = np.abs(stats.zscore(df[numeric_cols], nan_policy='omit'))
    outliers = (z_scores > 3).sum()
    print("Número de outliers por variable numérica:\n", outliers)

    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df[numeric_cols], orient="h", palette="coolwarm")
    plt.title("Distribución y outliers por variable numérica")
    plt.show()
else:
    print("⚠️ No hay columnas numéricas para detectar outliers.")


# 6️⃣ Mapa de correlaciones detallado
if len(numeric_cols) > 1:
    plt.figure(figsize=(10, 8))
    corr = df[numeric_cols].corr()
    sns.heatmap(corr, annot=True, cmap="RdBu_r", center=0, fmt=".2f")
    plt.title("Mapa de correlación (variables numéricas)")
    plt.show()
else:
    print("⚠️ No hay suficientes columnas numéricas para correlacionar.")


# 7️⃣ Análisis de correlaciones más relevantes
if len(numeric_cols) > 1:
    corr_sorted = corr.unstack().sort_values(ascending=False)
    print("🔍 Correlaciones más fuertes (positivas y negativas):")
    display(corr_sorted[(corr_sorted < 0.99) & (corr_sorted > 0.5) | (corr_sorted < -0.5)].dropna().head(10))


# 8️⃣ Relaciones entre variables categóricas y numéricas
cat_cols = df.select_dtypes(include=['object']).columns

if len(cat_cols) > 0 and len(numeric_cols) > 0:
    cat_to_plot = cat_cols[0]  # primera variable categórica como ejemplo
    num_to_plot = numeric_cols[0]  # primera variable numérica como ejemplo

    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x=cat_to_plot, y=num_to_plot, palette="Set3")
    plt.title(f"Relación entre {cat_to_plot} y {num_to_plot}")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("⚠️ No hay suficientes variables categóricas o numéricas para analizar relaciones.")


# 9️⃣ Análisis de frecuencia de variables categóricas principales
for col in cat_cols[:3]:  # mostramos las tres primeras
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index[:10], palette="crest")
    plt.title(f"Distribución de las 10 categorías principales en '{col}'")
    plt.xlabel("Frecuencia")
    plt.ylabel(col)
    plt.show()


# 🔟 Correlación entre variables textuales (longitud de texto)
if "job_description" in df.columns:
    df["desc_length"] = df["job_description"].astype(str).apply(len)
    plt.figure(figsize=(10, 6))
    sns.histplot(df["desc_length"], bins=30, color="orange", kde=True)
    plt.title("Distribución de longitud de descripciones de empleo")
    plt.xlabel("Longitud del texto")
    plt.ylabel("Frecuencia")
    plt.show()


# 1️⃣1️⃣ Conclusiones rápidas (en texto)
print("\n✅ EDA avanzado completado correctamente.")
print("📈 Se han generado distribuciones, correlaciones y análisis de outliers.")
print("💡 Revisa los gráficos para identificar patrones, valores atípicos y relaciones clave entre variables.")
```
