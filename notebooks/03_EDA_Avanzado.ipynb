```python
# ============================================================
# 03_EDA_Avanzado.ipynb ‚Äî Exploraci√≥n Avanzada de Datos (EDA)
# ============================================================

# 1Ô∏è‚É£ Importaci√≥n de librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuraci√≥n general
sns.set(style="whitegrid", palette="muted", font_scale=1.1)
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["axes.labelsize"] = 12

print("‚úÖ Librer√≠as importadas correctamente.")


# 2Ô∏è‚É£ Carga del dataset
df = pd.read_csv("../data/fusionado_reducido.csv")
print(f"\nüìä Dataset cargado correctamente con {df.shape[0]} filas y {df.shape[1]} columnas.\n")

# Vista previa
display(df.head())


# 3Ô∏è‚É£ An√°lisis general del dataset
print("----- INFORMACI√ìN GENERAL -----")
df.info()

print("\n----- VALORES NULOS -----")
print(df.isnull().sum().sort_values(ascending=False).head(10))

print("\n----- ESTAD√çSTICAS DESCRIPTIVAS NUM√âRICAS -----")
display(df.describe(include=[np.number]).T)

print("\n----- ESTAD√çSTICAS DESCRIPTIVAS CUALITATIVAS -----")
display(df.describe(include=['object']).T)


# 4Ô∏è‚É£ Distribuci√≥n de variables num√©ricas
numeric_cols = df.select_dtypes(include=[np.number]).columns

if len(numeric_cols) > 0:
    df[numeric_cols].hist(bins=25, figsize=(14, 10), color="skyblue", edgecolor="black")
    plt.suptitle("Distribuci√≥n de variables num√©ricas", fontsize=16)
    plt.show()
else:
    print("‚ö†Ô∏è No se detectaron columnas num√©ricas para graficar.")


# 5Ô∏è‚É£ Detecci√≥n y an√°lisis de outliers
if len(numeric_cols) > 0:
    print("\n----- DETECCI√ìN DE OUTLIERS (usando Z-Score) -----")
    z_scores = np.abs(stats.zscore(df[numeric_cols], nan_policy='omit'))
    outliers = (z_scores > 3).sum()
    print("N√∫mero de outliers por variable num√©rica:\n", outliers)

    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df[numeric_cols], orient="h", palette="coolwarm")
    plt.title("Distribuci√≥n y outliers por variable num√©rica")
    plt.show()
else:
    print("‚ö†Ô∏è No hay columnas num√©ricas para detectar outliers.")


# 6Ô∏è‚É£ Mapa de correlaciones detallado
if len(numeric_cols) > 1:
    plt.figure(figsize=(10, 8))
    corr = df[numeric_cols].corr()
    sns.heatmap(corr, annot=True, cmap="RdBu_r", center=0, fmt=".2f")
    plt.title("Mapa de correlaci√≥n (variables num√©ricas)")
    plt.show()
else:
    print("‚ö†Ô∏è No hay suficientes columnas num√©ricas para correlacionar.")


# 7Ô∏è‚É£ An√°lisis de correlaciones m√°s relevantes
if len(numeric_cols) > 1:
    corr_sorted = corr.unstack().sort_values(ascending=False)
    print("üîç Correlaciones m√°s fuertes (positivas y negativas):")
    display(corr_sorted[(corr_sorted < 0.99) & (corr_sorted > 0.5) | (corr_sorted < -0.5)].dropna().head(10))


# 8Ô∏è‚É£ Relaciones entre variables categ√≥ricas y num√©ricas
cat_cols = df.select_dtypes(include=['object']).columns

if len(cat_cols) > 0 and len(numeric_cols) > 0:
    cat_to_plot = cat_cols[0]  # primera variable categ√≥rica como ejemplo
    num_to_plot = numeric_cols[0]  # primera variable num√©rica como ejemplo

    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x=cat_to_plot, y=num_to_plot, palette="Set3")
    plt.title(f"Relaci√≥n entre {cat_to_plot} y {num_to_plot}")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("‚ö†Ô∏è No hay suficientes variables categ√≥ricas o num√©ricas para analizar relaciones.")


# 9Ô∏è‚É£ An√°lisis de frecuencia de variables categ√≥ricas principales
for col in cat_cols[:3]:  # mostramos las tres primeras
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index[:10], palette="crest")
    plt.title(f"Distribuci√≥n de las 10 categor√≠as principales en '{col}'")
    plt.xlabel("Frecuencia")
    plt.ylabel(col)
    plt.show()


# üîü Correlaci√≥n entre variables textuales (longitud de texto)
if "job_description" in df.columns:
    df["desc_length"] = df["job_description"].astype(str).apply(len)
    plt.figure(figsize=(10, 6))
    sns.histplot(df["desc_length"], bins=30, color="orange", kde=True)
    plt.title("Distribuci√≥n de longitud de descripciones de empleo")
    plt.xlabel("Longitud del texto")
    plt.ylabel("Frecuencia")
    plt.show()


# 1Ô∏è‚É£1Ô∏è‚É£ Conclusiones r√°pidas (en texto)
print("\n‚úÖ EDA avanzado completado correctamente.")
print("üìà Se han generado distribuciones, correlaciones y an√°lisis de outliers.")
print("üí° Revisa los gr√°ficos para identificar patrones, valores at√≠picos y relaciones clave entre variables.")
```
